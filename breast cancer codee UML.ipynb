## Appyling PCA & Displaying Images

```python
i = 0
for image in images.keys():
    img_copy = images[image].copy()

    gray_img = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)

    # Flatten the grayscale image into a 1D array
    flattened_img = gray_img.flatten()

    # Reshape the flattened image into a 2D array
    reshaped_img = np.reshape(flattened_img, (gray_img.shape[0] * gray_img.shape[1], 1))

    # Perform PCA on the reshaped image
    mean, eigenvectors = cv2.PCACompute(reshaped_img, mean=None, maxComponents=gray_img.shape[0])

    # Project the image onto the eigenvectors
    projected_img = cv2.PCAProject(reshaped_img, mean, eigenvectors)

    # Reshape the projected image back into the original image shape
    reconstructed_img = np.reshape(projected_img, gray_img.shape)

    print(images[image].shape)
    print(reconstructed_img.shape)


    # Display the original and reconstructed images
    cv2.imshow("Original Image", gray_img)
    cv2.imshow("Reconstructed Image", reconstructed_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    i += 1
    print("\n")
    if i>=4:
        break
```

The change in shape from (471, 562, 3) to (471, 562) suggests that you have performed dimensionality reduction using PCA (Principal Component Analysis) on an image with the original shape of (471, 562, 3). The first two dimensions of (471, 562) represent the height and width of the image after PCA, while the third dimension (3) is missing, indicating that the number of color channels has been reduced from 3 to 1

The computed image may be in high contrast because the PCA transformation can accentuate the differences between the high and low intensity values in the grayscale image. This can result in an image that appears to have higher contrast compared to the original. Additionally, the high contrast can also be due to the normalization that is performed as part of the PCA calculation, which can cause the intensity values to be scaled differently in the reconstructed image.

```python
import numpy as np
import cv2

for image in images.keys():
    # Read image
    data = images[image].copy()



    
    print(data.shape)


    
    data = cv2.cvtColor(data, cv2.COLOR_BGR2GRAY)


    print(data.shape)



    # Take the 2D DFT of the image
    f = np.fft.fft2(data)

    # Shift the zero-frequency component to the center of the spectrum
    fshift = np.fft.fftshift(f)

    # Calculate the magnitude spectrum
    magnitude_spectrum = 20*np.log(np.abs(fshift))

    # Apply a high-pass filter to remove low-frequency components
    rows, cols = data.shape
    crow, ccol = rows // 2, cols // 2
    mask = np.ones((rows, cols), np.uint8)
    mask[crow-30:crow+30, ccol-30:ccol+30] = 0
    fshift_filtered = fshift * mask

    # Calculate the filtered magnitude spectrum
    magnitude_spectrum_filtered = 20*np.log(np.abs(fshift_filtered))

    # Take the inverse DFT to get the filtered image
    img_back = np.fft.ifft2(np.fft.ifftshift(fshift_filtered)).real

    # Display the original and filtered images and magnitude spectra
    cv2.imshow('Original Image', data)
    cv2.imshow('Filtered Image', img_back.astype(np.uint8))
    cv2.imshow('Magnitude Spectrum', magnitude_spectrum.astype(np.uint8))
    cv2.imshow('Filtered Magnitude Spectrum', magnitude_spectrum_filtered.astype(np.uint8))
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
    
    break
```

## BUTTERWORTH BABYYYYYYYYY

```python
import cv2
import numpy as np

i=0

for image in images.keys():
     
    img = images[image].copy()

    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
    # Calculate the 2D FFT of the image
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)

    # Construct the Butterworth filter
    rows, cols = img.shape
    crow, ccol = rows // 2, cols // 2
    D = 50   # cutoff frequency
    n = 2    # filter order
    butterworth_filter = np.zeros((rows, cols), np.float32)
    for i in range(rows):
        for j in range(cols):
            dist = np.sqrt((i - crow) ** 2 + (j - ccol) ** 2)
            butterworth_filter[i, j] = 1 / (1 + (dist / D) ** (2 * n))

    # Apply the filter to the frequency domain
    filtered_fshift = fshift * butterworth_filter

    # Shift the zero-frequency component back to the top-left corner
    filtered_f = np.fft.ifftshift(filtered_fshift)

    # Calculate the inverse 2D FFT to get the filtered image
    filtered_img = np.fft.ifft2(filtered_f).real

    # Display the original and filtered images side by side
    cv2.imshow('Original Image', img)
    cv2.imshow('Butterworth Filtered Image', filtered_img.astype(np.uint8))
    cv2.waitKey(0)
    cv2.destroyAllWindows()



    break
```

# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from sklearn.decomposition import PCA

i = 0
for image in images.keys():
    # Load the image
    image = images[image].copy()
    # Convert the image to grayscale
    image_gray = np.dot(image[..., :3], [0.299, 0.587, 0.114])

    # Compute the size of the original image
    orig_size = image_gray.nbytes

    # Apply PCA
    pca = PCA(n_components=210)
    pca.fit(image_gray)

    # Extract PCA feature
    pca_feature = pca.transform(image_gray)

    # Compute the size of the PCA feature
    pca_size = pca_feature.nbytes

    # Reconstruct the image using the PCA feature
    image_reconstructed = pca.inverse_transform(pca_feature).reshape(image_gray.shape)

    # Compute the compression ratio
    compression_ratio = float(orig_size) / pca_size

    # Compute the signal-to-noise ratio (SNR)
    signal_power = np.sum(image_gray ** 2)
    noise_power = np.sum((image_gray - image_reconstructed) ** 2)
    snr = 10 * np.log10(signal_power / noise_power)

    # Plot the original and reconstructed images
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(image_gray, cmap='gray')
    ax[0].set_title('Original Image')
    ax[1].imshow(image_reconstructed, cmap='gray')
    ax[1].set_title('Reconstructed Image')
    plt.show()

    # Display the compression ratio and SNR
    print(f"Compression ratio: {compression_ratio:.2f}")
    print(f"SNR: {snr:.2f} dB")

    i += 1
    print("\n")
    if i>=4:
        break
```

By scaling the data, you may have reduced the influence of any outliers or features with a large scale that were affecting the clustering results.

In summary, normalizing the data with min-max scaling helped to remove the effect of feature scale, which could have affected the clustering results. This allowed K-means to identify the clusters more accurately, resulting in a more meaningful insight.
