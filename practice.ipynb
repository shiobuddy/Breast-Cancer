```python
import cv2
import numpy as np

# Load the image
img = cv2.imread(r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_benign\benign 3.jpg")

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding to segment the image
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

# Apply morphological operations to remove noise and fill holes
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
sure_bg = cv2.dilate(opening,kernel,iterations=3)
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)

# Apply watershed transform
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers+1
markers[unknown==255] = 0
markers = cv2.watershed(img,markers)
img[markers == -1] = [255,0,0]

# Display the output image
cv2.imshow('Output', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions

model = VGG16(weights='imagenet', include_top=True)

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = image.load_img(os.path.join(folder, filename), target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        images.append(x)
    return images

benign_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_benign"
malignant_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_malignant"
normal_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_normal"

benign_images = load_images_from_folder(benign_folder)
malignant_images = load_images_from_folder(malignant_folder)
normal_images = load_images_from_folder(normal_folder)

for i, images in enumerate([benign_images, malignant_images, normal_images]):
    print(f'Predictions for set {i+1}:')
    for img in images:
        preds = model.predict(img)
        print(decode_predictions(preds, top=1)[0][0][1])
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import EfficientNetB0, preprocess_input
from sklearn.metrics import accuracy_score

# Load the EfficientNetB0 model
model = EfficientNetB0(weights='imagenet', include_top=True)

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = image.load_img(os.path.join(folder, filename), target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        images.append(x)
    return images

benign_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_benign"
malignant_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_malignant"
normal_folder = r"C:\Users\Najeeb\Desktop\Sem 4\Breast cancer dataset\Dataset_BUSI_with_GT\Cleaned Breast Cancer Dataset\new_new_normal"

benign_images = load_images_from_folder(benign_folder)
malignant_images = load_images_from_folder(malignant_folder)
normal_images = load_images_from_folder(normal_folder)

# Define ground truth labels for the test images
benign_labels = np.zeros(len(benign_images))
malignant_labels = np.ones(len(malignant_images))
normal_labels = np.full(len(normal_images), 2)

# Concatenate the test images and labels
test_images = np.concatenate([benign_images, malignant_images, normal_images], axis=0)
test_labels = np.concatenate([benign_labels, malignant_labels, normal_labels], axis=0)

# Predict labels for test images
predicted_labels = []
for img in test_images:
    preds = model.predict(img)
    predicted_labels.append(np.argmax(preds))

# Calculate accuracy
accuracy = accuracy_score(test_labels, predicted_labels)
print(f"Accuracy: {accuracy}")
```

```python
# Set the batch size
batch_size = 32

# Convert benign and malignant lists to numpy arrays
benign = np.array(benign)
malignant = np.array(malignant)
normal = np.array(normal)

# Load the images with three channels
benign = np.repeat(benign, 3, axis=-1)
malignant = np.repeat(malignant, 3, axis=-1)
normal = np.repeat(normal, 3, axis=-1)

# Reshape the images to the required input shape of the model
benign = benign.reshape(len(benign), 224, 224, 3)
malignant = malignant.reshape(len(malignant), 224, 224, 3)
normal = normal.reshape(len(normal), 224, 224, 3)

# Extract features from the benign images
benign_features = []
for i in range(0, len(benign), batch_size):
    batch = benign[i:i+batch_size]
    features = extract_features(batch)
    benign_features.append(features)
benign_features = np.concatenate(benign_features)

# Extract features from the malignant images
malignant_features = []
for i in range(0, len(malignant), batch_size):
    batch = malignant[i:i+batch_size]
    features = extract_features(batch)
    malignant_features.append(features)
malignant_features = np.concatenate(malignant_features)
```

```python
normal_features = []
for i in range(0, len(normal), batch_size):
    batch = normal[i:i+batch_size]
    features = extract_features(batch)
    normal_features.append(features)
normal_features = np.concatenate(normal_features)
```

```python
# Concatenate the features and labels of all the images
features = np.concatenate((benign_features, malignant_features, normal_features))
labels = np.concatenate((labels_benign, labels_mali, labels_norm))
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Train a linear discriminant analysis model
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# Train a support vector machine model
svm = SVC(kernel='linear', C=1, probability=True)
svm.fit(X_train, y_train)
```

```python
# Evaluate the performance of the linear discriminant analysis model
lda_pred = lda.predict(X_test)
lda_acc = accuracy_score(y_test, lda_pred)
print("Linear Discriminant Analysis Accuracy: {:.2f}%".format(lda_acc*100))

# Evaluate the performance of the support vector machine model
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
print("Support Vector Machine Accuracy: {:.2f}%".format(svm_acc*100))
```

```python
framObjTrain = {'img' : [],
           'mask' : []
          }

## defining data Loader function
def LoadData( frameObj = None, imgPath = None, maskPath = None, shape = 256):
    imgNames = os.listdir(imgPath)
    
    names = []
    maskNames = []
    unames = []
    
    for i in range(len(imgNames)):
        unames.append(imgNames[i].split(')')[0])
    
    unames = list(set(unames))
    
    for i in range(len(unames)):
        names.append(unames[i]+').png')
        maskNames.append(unames[i]+')_mask.png')
    
    imgAddr = imgPath + '/'
    maskAddr = maskPath + '/'
    
    for i in range (len(names)):
        img = plt.imread(imgAddr + names[i])
        mask = plt.imread(maskAddr + maskNames[i])
        
        img = cv2.resize(img, (shape, shape)) 
        mask = cv2.resize(mask, (shape, shape))
        
        frameObj['img'].append(img)
        frameObj['mask'].append(mask)
        
    return frameObj
```
